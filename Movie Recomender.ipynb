# MOVIE RECOMMENDER SYSTEM

IMPORTING LIBRARIES

```python
import numpy as np
import pandas as pd
```

GETTING OUR DATA

```python
column_names = ['user_id', 'item_id', 'rating', 'timestamp']
df = pd.read_csv(r'C:\Users\HP\Desktop\MRS\u.data', sep='\t', names=column_names)
```

GETTING THE MOVIE TITLE FOR THE RESPECTIVE "item_id"

```python
movie_titles=pd.read_csv(r'C:/Users/HP/Desktop/MRS/Movie_Id_Titles')
movie_titles.head()
```

MERGING THE MOVIE TITLE TO THE DATA FRAME

```python
df=pd.merge(df,movie_titles,on='item_id')
df.head()
```

```python
ratings=pd.DataFrame(df.groupby('title')['rating'].mean())
ratings.head()
```

```python
ratings['Number Of Ratings']=pd.DataFrame(df.groupby('title')['rating'].count())
ratings.head()
```

```python
ratingmat=df.pivot_table(index='user_id',columns='title',values='rating')
ratingmat.head()
```

```python
ratings.sort_values('Number Of Ratings',ascending=False).head(10)
```

```python
ratings.head()
```

```python
fargo_ur=ratingmat['Fargo (1996)']
fargo_ur.head()
```

```python
sim_to_fargo=ratingmat.corrwith(fargo_ur)
sim_to_fargo.head()
```

```python
corr_fargo=pd.DataFrame(sim_to_fargo,columns=['Correlation'])
corr_fargo.dropna(inplace=True)
corr_fargo.head()
```

```python
corr_fargo.sort_values('Correlation',ascending=False).head(10)
```

```python
corr_fargo=corr_fargo.join(ratings['Number Of Ratings'])
corr_fargo.head()
```

```python
corr_fargo[corr_fargo['Number Of Ratings']>100].sort_values('Correlation',ascending=False).head()
```

```python
starwars_ur=ratingmat['Star Wars (1977)']
sim_to_starwars=ratingmat.corrwith(starwars_ur)
corr_starwars=pd.DataFrame(sim_to_starwars,columns=['Correlation'])
corr_starwars.dropna(inplace=True)
corr_starwars=corr_starwars.join(ratings['Number Of Ratings'])
corr_starwars[corr_starwars['Number Of Ratings']>100].sort_values('Correlation',ascending=False).head()
```
